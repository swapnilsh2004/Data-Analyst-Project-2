# Gender distribution

```python
plt.figure(figsize=(5, 5))
ax=sns.countplot(data=df, x="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
gb=df.groupby("ParentEduc").agg({"MathScore":"mean","WritingScore":"mean","ReadingScore":"mean"})
```

```python
gb
```

```python
sns.heatmap(gb,annot=True)
plt.title("Relation between student performance and parent eduction")
plt.show()
```

```python
#from the av=bove chart it has concluded that parent eduction impact on student eduction
```

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","WritingScore":"mean","ReadingScore":"mean"})
gb1
```

```python
sns.heatmap(gb1,annot=True)
plt.title("Relation between student performance and parent marital status")
plt.show()
```

```python
#from above chart it has conclude that there is no impact of parent marital status on student eduction
```

# Distribution of ethnics group

```python
groupA = df.loc[df["EthnicGroup"] == "group A"].count()
groupB = df.loc[df["EthnicGroup"] == "group B"].count()
groupC = df.loc[df["EthnicGroup"] == "group C"].count()
groupD = df.loc[df["EthnicGroup"] == "group D"].count()
groupE = df.loc[df["EthnicGroup"] == "group E"].count()
```

```python
l=["GroupA","GroupB","GroupC","GroupD","GroupE"]
mylist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mylist,labels=l,autopct="%.2f%%")
plt.title("Distribution of Ethincs Group")
plt.show()
```
